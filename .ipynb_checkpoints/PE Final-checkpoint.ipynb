```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def diff(dataframe,x,y): #função para calcular a derivada
    '''Retorna a derivada dy/dx de tal dataframe.\nx e y são strings.'''
    pontos=[0]
    for i in range(1,len(dataframe[x])):
        try:
            pontos.append((dataframe[y][i]-dataframe[y][i-1])/(dataframe[x][i]-dataframe[x][i-1]))
        except IndexError:
            continue
    return pontos
    
def pe(dataframe,n_pontos,plotar=False):
    '''Retorna o ponto de equivalência da titulação (Vtit e pH). O dataframe deve conter uma coluna com 'pH' e outra com 'Vtit'.'''
    assert n_pontos==1 or n_pontos== 2 or n_pontos== 3,('Não se titula mais de 3 pontos, duá.')
    
    dataframe=dataframe[['pH','Vtit']] #tirando as colunas vazias (frescura)
    dataframe=dataframe.assign(diff=diff(dataframe,'Vtit','pH')) #adicionando uma coluna com a derivada
    
    if n_pontos==1:
        pH_eq=dataframe.loc[dataframe['diff'] == dataframe['diff'].max()]['pH'] #bagulhada pra tirar o pH a partir do valor da maior derivada
        Vtit_eq=dataframe.loc[dataframe['diff'] == dataframe['diff'].max()]['Vtit'] #mesma coisa pro Vtit
        print('Vtit='+str(Vtit_eq.values[0])+'\npH='+str(pH_eq.values[0])) #.values converte  o dataframpe para lista e o [0] pega o primeiro (e único) valor como float
        
        if plotar==True:  
            plt.plot(dataframe['Vtit'],dataframe['diff'],color='black')
    else:
        #fazendo o clustering
        km=KMeans(n_clusters=n_pontos)
        diff_pred=km.fit_predict(dataframe[['pH','Vtit']])
        dataframe['Cluster']=diff_pred
        
        if n_pontos==2:
            df1=dataframe[dataframe.Cluster==0]
            df2=dataframe[dataframe.Cluster==1]
            
            pH_eq1=dataframe.loc[dataframe['diff'] == df1['diff'].max()]['pH']
            Vtit_eq1=dataframe.loc[dataframe['diff'] == df1['diff'].max()]['Vtit']
            
            pH_eq2=dataframe.loc[dataframe['diff'] == df2['diff'].max()]['pH']
            Vtit_eq2=dataframe.loc[dataframe['diff'] == df2['diff'].max()]['Vtit']
            
            print('PE 1:'+'\nVtit='+str(Vtit_eq1.values[0])+'\npH='+str(pH_eq1.values[0]))
            print('\nPE 2:'+'\nVtit='+str(Vtit_eq2.values[0])+'\npH='+str(pH_eq2.values[0]))
            
            if plotar==True:
                plt.plot(df1['Vtit'],df1['diff'],color='red')
                plt.plot(df2['Vtit'],df2['diff'],color='black')
        else:
            df1=dataframe[dataframe.Cluster==0]
            df2=dataframe[dataframe.Cluster==1]
            df3=dataframe[dataframe.Cluster==2]
            
            pH_eq1=dataframe.loc[dataframe['diff'] == df1['diff'].max()]['pH']
            Vtit_eq1=dataframe.loc[dataframe['diff'] == df1['diff'].max()]['Vtit']
            
            pH_eq2=dataframe.loc[dataframe['diff'] == df2['diff'].max()]['pH']
            Vtit_eq2=dataframe.loc[dataframe['diff'] == df2['diff'].max()]['Vtit']
            
            pH_eq3=dataframe.loc[dataframe['diff'] == df3['diff'].max()]['pH']
            Vtit_eq3=dataframe.loc[dataframe['diff'] == df3['diff'].max()]['Vtit']            
            
            print('PE 1:'+'\nVtit='+str(Vtit_eq1.values[0])+'\npH='+str(pH_eq1.values[0]))
            print('\nPE 2:'+'\nVtit='+str(Vtit_eq2.values[0])+'\npH='+str(pH_eq2.values[0]))
            print('\nPE 3:'+'\nVtit='+str(Vtit_eq3.values[0])+'\npH='+str(pH_eq3.values[0]))
            
            if plotar==True:
                plt.plot(df1['Vtit'],df1['diff'],color='red')
                plt.plot(df2['Vtit'],df2['diff'],color='black')  
                plt.plot(df2['Vtit'],df2['diff'],color='green')  
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
df=pd.read_csv('Projeto - Gustavo e cia..csv',sep=';',decimal=',')
pe(df,2,True)
```
